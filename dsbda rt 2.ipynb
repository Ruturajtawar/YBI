```python
Q1 = df1['math score'].quantile(0.25)
Q3 = df1['math score'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
filtered_df = df1[(df1['math score'] >= lower_bound) & (df1['math score'])]
filtered_df
```

```python
sn.boxplot(y=filtered_df['math score'])
plt.show()
```

```python
Q1 = df1['math score'].quantile(0.25)
Q3 = df1['math score'].quantile(0.75)
median = df1['math score'].median()
threshold = 1.5 * (Q3-Q1)
lower_bound = median - threshold
upper_bound = median + threshold
df1.loc[df1['math score'] < lower_bound,'math score'] = median
df1.loc[df1['math score'] > upper_bound,'math score'] = median
df1
```

```python
plt.title("Filtered Histogram")
sn.histplot(filtered_df['math score'],bins=30, color='blue')
plt.xlabel("math score")
plt.ylabel('Frequency')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.title("Original Histogram")
sn.histplot(filtered_df['math score'],bins=30, kde=True)
plt.subplot(1,2,2)
plt.title("Filtered Histogram")
sn.histplot(filtered_df['math score'],bins=30, kde=True)
plt.tight_layout()
plt.show()
```
